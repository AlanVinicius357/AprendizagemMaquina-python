```python
import pandas as pd
```

```python
diabetes_df = pd.read_csv('diabetes_clean.csv')

diabetes_df.head(5)
```

```python
diabetes_df['glucose']
```

```python
diabetes_df[diabetes_df['glucose'] > 0 ]
```

```python
diabetes_df[diabetes_df['bmi'] > 0]
```

```python
diabetes_df = diabetes_df[
    (diabetes_df['glucose'] > 0)
    & (diabetes_df ['bmi'] > 0)
]
```

```python
#todas colunas - glucose
#pregnancies | distolic | triceps | insuline | bmi ...
X = diabetes_df.drop("glucose", axis=1).values

#apenas a coluna glucose
y = diabetes_df["glucose"].values
print(type(X), type(y))
```

```python
X_bmi = X[:, 4]
print(y.shape, X_bmi.shape)
```

```python
# -1 é equivalente a última linha, e 1 afirma que temos apenas uma única coluna.  
X_bmi = X_bmi.reshape(-1, 1)  # cria uma matriz 2D de uma única coluna.
print(y.shape, X_bmi.shape)
```

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_bmi, y)

plt.ylabel("Blood Glucose (mg/dl)")
plt.xlabel("Body Mass Index")

plt.show()
```
